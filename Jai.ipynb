```python
movies=movies[['id','original_title','genres','overview']]
```

```python
movies['tags']=movies['overview']+movies['genres']
```

```python
movies
```

```python
new_data=movies.drop(columns=['overview','genres'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=1287,stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['original_title']=='Star Wars: The Force Awakens'].index[0]
```

```python
import pandas as pd
import numpy as np
distance = sorted(list(enumerate(similarity[7])), key=lambda x: x[1])
middle_index = len(distance) // 7
if len(distance) % 7 == 1:
    print(new_data.iloc[distance[middle_index][0]].original_title)
else:
    print(new_data.iloc[distance[middle_index][0]].original_title)
```

```python
def recommend(movies):
    if movies in new_data['original_title'].values:
        try:
            index = new_data[new_data['original_title'] == movies].index[0]
            distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])[1:6]

            print(f"Top 5 recommended movies based on '{movies}':")
            for i in distance:
                print(new_data.iloc[i[0]]['original_title'])
        except IndexError:
            print("Movie not found in the dataset.")
        except Exception as e:
            print(f"An error occurred: {e}")
    else:
        print(f"'{movies}' not found in the dataset. Please check the movie title.")
```

```python
recommend("Insurgent")
```
